```python
df = pd.read_csv("/kaggle/input/diamonds-prices/Diamonds Prices2022.csv")
df.head(2)
```

```python
df.drop(columns = ["Unnamed: 0"],inplace = True)
```

```python
df.shape
```

```python
df["price"].agg(["min","max","mean","median"])
```

```python
x = df.select_dtypes("object")
x
```

```python
x.apply(lambda x:x.unique())
```

```python
df["cut"] = df["cut"].replace({"Ideal":3,"Premium":4,"Good":1,"Very Good":2,"Fair":0})
```

```python
df["cut"].unique()
```

```python
df_final  = pd.get_dummies(df,drop_first = True)
df_final.head(2)
```

```python
X = df_final.drop(columns = "price")
y = df_final["price"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.10,random_state = 1)
```

```python
X_train.columns
```

```python
X_train.shape
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
dtr = DecisionTreeRegressor()
dtr.fit(X_train,y_train)
```

```python
ypred = dtr.predict(X_test)
```

```python
from sklearn.metrics import r2_score,mean_squared_error
```

```python
r2_score(y_test,ypred)
```

```python
from sklearn.tree import plot_tree
```

```python
x = plot_tree(dtr,filled = True)
x
```

```python
dtr.feature_importances_
```

```python
df_final.head(2)
```

```python
from sklearn.ensemble import RandomForestRegressor 
```

```python
rfr = RandomForestRegressor(max_depth = 5)
rfr.fit(X_train,y_train)
```

```python
y_pred = rfr.predict(X_test)
```

```python
r2_score(y_test,y_pred)
```

```python
np.sqrt(mean_squared_error(y_test,y_pred))
```

```python
#when performance of the model is less than desired , so we can go for different techniques to solve this issue 
# 1. feature engineering 
# 2.feature selection 
# 3.hyperparameter tuning
# 4. optimization 

#when performance is overachieved , so we can do the following :
# 1. Regularization 
# 2. cross validation 
```

```python
from sklearn.feature_selection import SelectKBest,chi2,f_regression
```

```python
X_train.shape
```

```python
X_new = SelectKBest(f_regression,k=10).fit_transform(X,y)
```

```python
X_new.shape
```

```python
X_train,X_test,y_train,y_test = train_test_split(X_new,y,test_size = 0.10,random_state = 1)
```

```python
rfr = RandomForestRegressor()
rfr.fit(X_train,y_train)
```

```python
y_pred = rfr.predict(X_test)
```

```python
r2_score(y_test,y_pred)
```

```python
#hyperparameter tuning 
#1 gridsearch 
#2. randomsearch 
```

```python
param_grid = {"n_estimators":[50,60,70,80,90,100],"criterion": ["squared_error", "absolute_error", "poisson"],"max_depth":[5,6,7,8,9,10]}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
grid = GridSearchCV(rfr,param_grid = param_grid,cv=5)
```

```python
grid
```

```python
grid.fit(X_train,y_train)
```
